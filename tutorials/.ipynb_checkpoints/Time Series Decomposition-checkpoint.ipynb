```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from typing import List, Tuple
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from statsforecast import StatsForecast
from statsforecast.models import ARIMA, AutoARIMA
from statsforecast.arima import arima_string
import warnings
warnings.filterwarnings('ignore')
```

# Create a dataset that has seasonalities, trend and noise

```python
# Set random seed
np.random.seed(42)

# AR(3) parameters
ar_params = [0.2, -0.01, 0.02]
n = 300
noise = np.random.normal(0, 0.1, size=n)

# Generate AR(3) base
y = [0, 0, 0]
for t in range(3, n):
    val = (
        ar_params[0] * y[t - 1] +
        ar_params[1] * y[t - 2] +
        ar_params[2] * y[t - 3] +
        noise[t]
    )
    y.append(val)

# Add linear trend
trend = 0.01 * np.arange(n)
y_trended = np.array(y) + trend

# üîÅ Add single seasonal component
seasonal_period = 24  # e.g. seasonality every 50 steps
seasonality = 0.3 * np.sin(2 * np.pi * np.arange(n) / seasonal_period)

# Add to signal
y_final = y_trended + seasonality

# Create DataFrame
df = pd.DataFrame({'value': y_final})
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df['value'])
```
